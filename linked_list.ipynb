```python
class LinkedList:
    # FUNCTION TO INITIALZE HEAD
    
    def __init__(self):
        self.head = None
        
    # FUNCTION TO INSERT A NEW
    # NODE AT THE BEGINNING
    
    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node
        
    # UTILITY FUNCTION TO PRINT IT
    # THE LINKED LINKEDLIST
    
    def printList(self):
        temp = self.head
        while(temp):
            print(temp.data, end=" ")
            temp = temp.next
    def detectLoop(self):
        s = set()
        temp = self.head
        while(temp):
            if (temp in s):
                return True
            s.add(temp)
            temp = temp.next
        return False
        
    
```

```python
# HOW MANY TIME THE ONE NODE COME IN LIST CAN BE COUNT BY FOLLOWINg
class node:
    def __init__(self,data):
        self.data = data
        self.next = None
class linkedlist:
    def __init__(self):
        self.head = None
    def push(self,new_node):
        new_node = node(new_node)
        new_node.next = self.head 
        self.head = new_node
    def count_one_node(self,that_node):
        current = self.head
        count = 0
        while(current is not None):
            if current.data == that_node:
                count += 1
            current = current.next
            return count
    def printlist(self):
        temp = self.head
        while(temp):
            print(temp.data , end=" ")
            temp = temp.next
my_list = linkedlist()
my_list.push(1)
my_list.push(1)
print("the list is")
my_list.printlist()
print()
print("value ",my_list.count_one_node(1))
for i in range(0,7):
    my_list.push(i)
    print()
    my_list.printlist()
print()
print("number of time one come" ,my_list.count_one_node(1))
```
